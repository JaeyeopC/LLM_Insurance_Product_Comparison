```python
# Cell: Refined HTML-to-Markdown Conversion with Encoding Detection
import os
from typing import Dict

# Install chardet if not available: pip install chardet
import chardet

try:
    from langchain_community.document_transformers import Html2TextTransformer
except ImportError as e:
    raise ImportError(
        "html2text package not found. Please install it:\n\n"
        "  pip install html2text\n\n"
        "Also ensure 'langchain_community' is installed or accessible in your environment.\n\n"
        f"Original Error: {e}"
    )

from langchain.schema import Document

def transform_html_to_markdown(input_dir: str, output_dir: str) -> Dict[str, str]:
    """
    Recursively processes all .html files in `input_dir`,
    detects the best possible file encoding, converts them to Markdown,
    and saves them to `output_dir` (preserving subdirectory structure).
    
    Returns a dictionary mapping the output file paths to their markdown content.
    """
    html2text = Html2TextTransformer()
    cleansed_text = {}

    for root, dirs, files in os.walk(input_dir):
        # Determine subdirectory path relative to input_dir
        relative_path = os.path.relpath(root, input_dir)
        # Create matching subdirectory in output_dir
        output_subdir = os.path.join(output_dir, relative_path)
        os.makedirs(output_subdir, exist_ok=True)

        for filename in files:
            if filename.endswith(".html"):
                input_path = os.path.join(root, filename)

                try:
                    # Read raw bytes
                    with open(input_path, 'rb') as f:
                        raw_content = f.read()
                    
                    # Detect encoding via chardet
                    detected = chardet.detect(raw_content)
                    detected_encoding = detected.get("encoding", "utf-8")

                    # Decode with the detected encoding; fallback to utf-8
                    try:
                        html_content = raw_content.decode(detected_encoding, errors='replace')
                    except:
                        html_content = raw_content.decode('utf-8', errors='replace')

                    # Convert HTML to Markdown via Html2TextTransformer
                    document = Document(page_content=html_content, metadata={"source": filename})
                    transformed_doc = html2text.transform_documents([document])[0]
                    markdown_content = transformed_doc.page_content

                    # Create the .md filename
                    markdown_filename = filename.replace(".html", ".md")
                    markdown_path = os.path.join(output_subdir, markdown_filename)

                    # Write the Markdown file
                    with open(markdown_path, 'w', encoding='utf-8') as md_file:
                        md_file.write(markdown_content)

                    cleansed_text[markdown_path] = markdown_content
                    print(f"Transformed and saved: {markdown_path} (encoding={detected_encoding})")

                except Exception as err:
                    print(f"Error processing file {filename} in {root}: {err}")

    return cleansed_text

# ------------------------------
# MAIN EXECUTION
# ------------------------------
# Convert HTML in "02_binary_products_v1" => "03_textbinary_products_v1"
markdown_results_v1 = transform_html_to_markdown(
    input_dir="02_binary_products_v1",
    output_dir="03_textbinary_products_v1"
)

# Convert HTML in "02_binary_products_v2" => "03_textbinary_products_v2"
markdown_results_v2 = transform_html_to_markdown(
    input_dir="02_binary_products_v2",
    output_dir="03_textbinary_products_v2"
)
```
