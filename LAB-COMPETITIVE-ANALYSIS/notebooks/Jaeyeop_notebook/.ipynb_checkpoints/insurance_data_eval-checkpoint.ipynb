```python
import pandas as pd
comparison_df = pd.read_excel("product_comparisons_table.xlsx") 
comparison_df
```

```python
golden_set = (
    comparison_df[['company_details', 'comparison']]
    .rename(columns={'company_details': 'query', 'comparison': 'expected_response'})
    .to_dict("records")
)
golden_set
```

```python
print(golden_set[0]['query'])
```

```python
print(golden_set[0]['expected_response'])
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables and initialize clients
load_dotenv()
OpenAI_key = os.getenv("OPENAI_API_KEY")  
Huggingface_key = os.getenv("HUGGINGFACE_API_KEY")

if not OpenAI_key or not Huggingface_key:
    raise ValueError("OPENAI_API_KEY or HUGGINGFACE_API_KEY is not set in the environment variables.")
else:
    print(OpenAI_key)
    print(Huggingface_key)
```

```python
from trulens.providers.huggingface import Huggingface
from trulens.providers.openai import OpenAI
from trulens.feedback import GroundTruthAgreement

hug_provider = Huggingface(api_key=Huggingface_key)  # for groundedness_measure_with_nli
provider = OpenAI(model_engine="gpt-4o")
ground_truth_collection = GroundTruthAgreement(golden_set, provider=provider)
gta = GroundTruthAgreement(golden_set, provider=provider)

similarity_score_list = []
comprehensiveness_score_list = []
groundedness_score_list = []
company_list = []
category_list = []

for i in range(len(comparison_df)):
    try:
        similarity_score = gta.agreement_measure(
            prompt=comparison_df['company_details'][i],
            response=comparison_df['comparison'][i]
        )[0]
    except Exception as e:
        print(f"An error occurred while processing similarity score at row {i}: {e}")
        print(f"comparison result:\n{comparison_df['company_names'][i]}")
        print(f"company details:\n{comparison_df['company_details'][i]}")
        similarity_score = 0

    try:
        comprehensiveness_score = provider.comprehensiveness_with_cot_reasons(
            source=comparison_df['company_details'][i],
            summary=comparison_df['comparison'][i],
        )[0]
    except Exception as e:
        print(f"An error occurred while processing comprehensiveness score at row {i}: {e}")
        print("comprehensiveness_score is set to 0")
        print(f"comparison result:\n{comparison_df['company_names'][i]}")
        print(f"company details:\n{comparison_df['company_details'][i]}")
        comprehensiveness_score = 0

    try:
        groundedness_score = provider.groundedness_measure_with_cot_reasons(
            source=comparison_df['company_details'][i],
            statement=comparison_df['comparison'][i],
        )[0]
    except Exception as e:
        print(f"An error occurred while processing groundedness score at row {i}: {e}")
        print("groundedness_score is set to 0")
        print(f"comparison result:\n{comparison_df['company_names'][i]}")
        print(f"company details:\n{comparison_df['company_details'][i]}")
        groundedness_score = 0
        
    print(f'comparison between {comparison_df["company_names"][i]} in {comparison_df["category"][i]}')
    print(f'similarity_score: {similarity_score}, comprehensiveness_score: {comprehensiveness_score}, groundedness_score: {groundedness_score}')    

    similarity_score_list.append(similarity_score)
    comprehensiveness_score_list.append(comprehensiveness_score)
    groundedness_score_list.append(groundedness_score)
    company_list.append(comparison_df['company_names'][i])
    category_list.append(comparison_df['category'][i])

result_df = pd.DataFrame({
    'similarity_score': similarity_score_list,
    'comprehensiveness_score': comprehensiveness_score_list,
    'groundedness_score': groundedness_score_list,
    'company_list': company_list,
    'category': category_list
})
```

```python
result_df
```

```python
result_df = result_df.fillna(0)
```

```python
import matplotlib.pyplot as plt

# Plotting the bar graph horizontally
result_df.set_index('category')[['similarity_score', 'comprehensiveness_score', 'groundedness_score']].plot(kind='barh', figsize=(12, 8))
plt.ylabel('Category')
plt.xlabel('Scores')
plt.title('Comparison Results')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = result_df[['groundedness_score', 'comprehensiveness_score', 'similarity_score']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Groundedness, Comprehensiveness, and Similarity Scores')
plt.show()
```

```python
import plotly.express as px

# Plotting the histogram for groundedness, comprehensiveness, and similarity scores using Plotly
fig = px.histogram(result_df, x=['groundedness_score', 'comprehensiveness_score', 'similarity_score'], 
                   barmode='overlay', 
                   title='Groundedness, Comprehensiveness, and Similarity Scores',
                   labels={'value': 'Scores', 'variable': 'Metrics'},
                   template='plotly_white',
                   nbins=100)  # Increased the number of bins

fig.update_layout(
    xaxis_title='Scores',
    yaxis_title='Count',
    legend_title='Metrics',
    legend=dict(x=1, y=0.5, traceorder='normal', font=dict(size=10)),
    width=600  # Reduced the width of the plot
)

fig.show()
```
